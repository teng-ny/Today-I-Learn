## 분포에 따라 함수그리기

```python
import scipy as sp
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pylab as plt
%matplotlib inline
```

```python
#정규분포 객체 생성
X=sp.stats.norm()

fig, axes=plt.subplots(1, 3, figsize=(12,3))
x_min_999, x_max_999=X.interval(0.999)
x999=np.linspace(x_min_999, x_max_999, 1000)

#plot
#PDF : 확률밀도함수
axes[0].plot(x999, X.pdf(x999), label="PDF")

#CDF : 누적분포함수 / SF : 생존함수=1-누적분포함수
axes[1].plot(x999, X.cdf(x999), label="CDF")
axes[1].plot(x999, X.sf(x999), label="SF")

#PPF : 누적분포함수 역함수
axes[2].plot(x999, X.ppf(x999), label="PPF")

axes[0].legend()
axes[1].legend()
axes[2].legend()
```

```python
#F분포(dfs2, 50)
X=sp.stats.f(2,50)

fig, axes=plt.subplots(1, 3, figsize=(12,3))
x_min_999, x_max_999=X.interval(0.999)
x999=np.linspace(x_min_999, x_max_999, 1000)

#plot
#PDF : 확률밀도함수
axes[0].plot(x999, X.pdf(x999), label="PDF")

#CDF : 누적분포함수 / SF : 생존함수=1-누적분포함수
axes[1].plot(x999, X.cdf(x999), label="CDF")
axes[1].plot(x999, X.sf(x999), label="SF")

#PPF : 누적분포함수 역함수
axes[2].plot(x999, X.ppf(x999), label="PPF")

axes[0].legend()
axes[1].legend()
axes[2].legend()
```

```python
#포아송분포(lambda5)
X=sp.stats.poisson(5)

fig, axes=plt.subplots(1, 3, figsize=(12,3))
x_min_999, x_max_999=X.interval(0.999)
x999=np.linspace(x_min_999, x_max_999, 1000)

#plot
#PMF : 확률질량함수
axes[0].plot(x999, X.pmf(x999), label="PMF")

#CDF : 누적분포함수 / SF : 생존함수=1-누적분포함수
axes[1].plot(x999, X.cdf(x999), label="CDF")
axes[1].plot(x999, X.sf(x999), label="SF")

#PPF : 누적분포함수 역함수
axes[2].plot(x999, X.ppf(x999), label="PPF")

axes[0].legend()
axes[1].legend()
axes[2].legend()
```

## 랜덤변수 및 분포

```python
#랜덤 변수의 랜덤표본이 분포와 일치하는지 확인
#df=5인 카이제곱
#1.분포에서 500개 샘플링
X=sp.stats.chi2(df=5)
X_samples=X.rvs(500)
```

```python
#2. 샘플링을 카이제곱 분포에 fit시킴
#모수y를 사용하여 적합 카이제곱 분포 생성
df, loc, scale=sp.stats.chi2.fit(X_samples)
Y=sp.stats.chi2(df=df, loc=loc, scale=scale)
```
