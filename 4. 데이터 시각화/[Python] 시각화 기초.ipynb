## figure 생성

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plt.figure() : 새로운 figure를 생성
#plt.show() : 생성된 모든 figure를 보여줌
fig=plt.figure()
plt.show()
```

```python
#figure크기 조정
#최초 차ㅏㅇ의 크기를 가로 세로 9인치로 설정
fig=plt.figure(figsize=(9,9))
```

## axes 생성
- axese는 figure내에서 축을 가지는 하나의 좌표평면
- 실제로 데이터가 그려지는 곳은 axes
- add_axes()메서드를 쓰면 해당 figure클래스에 axes를 생성할 수 있지만 add_subplot()을 사용해보려고 한다!

```python
fig=plt.figure()
#전체창을 가로 1칸, 세로 1칸으로 쪼개고 그 중 첫번째 칸에 ax라는 이름의 axes를 생성
ax=fig.add_subplot(1,1,1)
plt.show()
```

```python
fig=plt.figure()
#전체 창을 가로 1칸, 세로 2칸으로 쪼개고, 그 중 첫번째 칸에 ax1, 두번째 칸에 ax2를 생성
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)
plt.show()
```

## axes 설정
- axes마다 형식을 지정해줄 수 있음

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

#set메서드를 사용하여 x축, y축의 최댓값, 최솟값을 설정/그래프 제목, x축y축 이름을 설정
ax.set(xlim=[0., 1.], ylim=[-0.5, 2.5], title='Example', xlabel='xAxis', ylabel='yAxis')
plt.show()
```

```python
#따로따로 지정할 수도 있음
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.set_xlim([0., 1.])
ax.set_ylim([-0.5, 2.5])
ax.set_title("Example")
ax.set_xlabel("xAxis")
ax.set_ylabel("yAxis")
plt.show()
```

```python
# 글자크기도 설정 가능
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.set_xlim([0., 1.])
ax.set_ylim([-0.5, 2.5])
ax.set_title("Example", size=20)
ax.set_xlabel("xAxis", size=10)
ax.set_ylabel("yAxis", size=15)
plt.show()
```

## 그래프 그리기

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.set(xlim=[0., 7.], ylim=[0., 50.])

x=np.array([1,3,6,4])
y=np.array([10,20,25,40])

ax.plot(x,y)
plt.show()
```

```python
#선의 형식도 지정 가능
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.set(xlim=[0., 7.], ylim=[0., 50.])

x=np.array([1,3,6,4])
y=np.array([10,20,25,40])

ax.plot(x,y, color='red', linewidth=3)
plt.show()
```

```python
#scatterplot
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

x=np.array([0.3, 3.8, 1.2, 2.5])
y=np.array([11, 25, 9, 26])

ax.scatter(x,y, marker='^', color='darkgreen')
plt.show()
```
